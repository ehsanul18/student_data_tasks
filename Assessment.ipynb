```python
x_values = ['school', 'sex', 'address', 'Pstatus']

merge_ini_math_encoded = merge_ini_math

le = LabelEncoder()

for i in x_values:
    merge_ini_math_encoded[i] = le.fit_transform(merge_ini_math_encoded[i])
```

```python
merge_ini_math.Pstatus.value_counts()
```

```python
from sklearn.linear_model import LassoLars

ri = LassoLars(alpha=.1)

X = merge_ini_math_encoded[['school', 'sex', 'address', 'Pstatus', 'age', 'Medu', 'Fedu']]
y = merge_ini_math_encoded['G3']

ri.fit(X, y)
```
